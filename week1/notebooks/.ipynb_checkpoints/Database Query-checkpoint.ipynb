## Question 3. Count records 

```python
query_3 = """select count(*) from yellow_taxi_trips
where tpep_pickup_datetime >= timestamp '2021-01-15'
and tpep_pickup_datetime < timestamp '2021-01-16' """

cur.execute(query_3)
cur.fetchall()
```

## Question 4. Average

```python
query_4 = """select tip_amount,tpep_pickup_datetime
from yellow_taxi_trips
where tip_amount = (select max(tip_amount) 
from yellow_taxi_trips where tpep_pickup_datetime >= timestamp '2021-01-01'
and tpep_pickup_datetime < timestamp '2021-02-01')
"""

cur.execute(query_4)
cur.fetchall()
```

## Question 5. Most popular destination

```python
query_5 = """select "Zone", count(*) 
from yellow_taxi_trips
inner join zones on "DOLocationID" = zones."LocationID"
where "PULocationID" = (select "LocationID" from zones where "Zone" = 'Central Park') and
tpep_pickup_datetime >= timestamp '2021-01-14'
and tpep_pickup_datetime < timestamp '2021-01-15'
group by "Zone"
order by count desc"""

cur.execute(query_5)
cur.fetchall()
```

## Question 6. Most expensive route

```python
query_6 = """select po_z."Zone", do_z."Zone" as do_id, avg(total_amount) as avg_amnt
from yellow_taxi_trips
inner join zones po_z on "PULocationID"=po_z."LocationID"
inner join zones do_z on "DOLocationID"=do_z."LocationID"
group by po_z."Zone",do_z."Zone"
order by avg_amnt desc
"""

cur.execute(query_6)
cur.fetchall()
```
